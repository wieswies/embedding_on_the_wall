## 1. Data preparation

```python
embeddings = pd.read_parquet('../datasets/embeddings/embedded_politicians_pca_tsne.parquet')
```

```python
print(f'The dataset containing unique embeddings\nof both politicians and unknowns\nis of shape {embeddings.shape}\nand contains the columns:\n{embeddings.columns.tolist()}')
```

```python
label_encoder = LabelEncoder()
labels = embeddings['dataset']
y = label_encoder.fit_transform(labels)
embeddings['dataset_encoded'] = y
```

## 2. SVM development
### a. Split the data into training (only politicians), development (politicians + unknowns), and testing (politicians + unknowns)

```python
pol_data = embeddings[embeddings['dataset'] != 'Unknown']
unknown_data = embeddings[embeddings['dataset'] == 'Unknown']

# Split pol_data into training and development sets
train_data, dev_data = train_test_split(pol_data, test_size = 0.2, random_state = 2)

# Combine dev_data with unknown_data
dev_data = pd.concat([dev_data, unknown_data])

# Split dev_data into development and test sets
dev_data, test_data = train_test_split(dev_data, test_size = 0.5, random_state = 2)

# Check the sizes of the splits
print(f"Training data size: {train_data.shape}")
print(f"Development data size: {dev_data.shape}")
print(f"Test data size: {test_data.shape}")
```

```python
# Set X_train, y__train, X_dev, y_dev, X_test, y_test for the different data dimensionalities
X_train = np.vstack(train_data['embedding'].values)
X_train_PCA = np.vstack(train_data['PCA_64D'].values)
X_train_tSNE = np.vstack(train_data['tSNE_2D'].values)
y_train = train_data['dataset_encoded']

X_dev = np.vstack(dev_data['embedding'].values)
X_dev_PCA = np.vstack(dev_data['PCA_64D'].values)
X_dev_tSNE = np.vstack(dev_data['tSNE_2D'].values)
y_dev = dev_data['dataset_encoded']

X_test = np.vstack(test_data['embedding'].values)
X_test_PCA = np.vstack(test_data['PCA_64D'].values)
X_test_tSNE = np.vstack(test_data['tSNE_2D'].values)
y_test = test_data['dataset_encoded']
```

```python
X_train_data = [
    {'num_dim': 512, 'embeddings_array': X_train},
    {'num_dim': 64, 'embeddings_array': X_train_PCA},
    {'num_dim': 2, 'embeddings_array': X_train_tSNE}
]

X_dev_data = [
    {'num_dim': 512, 'embeddings_array': X_dev},
    {'num_dim': 64, 'embeddings_array': X_dev_PCA},
    {'num_dim': 2, 'embeddings_array': X_dev_tSNE}
]

X_test_data = [
    {'num_dim': 512, 'embeddings_array': X_test},
    {'num_dim': 64, 'embeddings_array': X_test_PCA},
    {'num_dim': 2, 'embeddings_array': X_test_tSNE}
]
```

### b. Explore thresholds

```python
print(f'Number of classes in the training set: {y_train.nunique()} ({sorted(y_train.unique())})\nNumber of classes in the dev set: {y_dev.nunique()} ({sorted(y_dev.unique())})')
```

```python
pipe_svc = Pipeline([('scaler', StandardScaler()), ('svc', SVC(probability=True))])
pipe_svc.fit(X_train, y_train)
```

```python
probabilities = pipe_svc.predict_proba(X_dev)
```

```python
from methods_model_development import convert_probabilities, max_probability_info

probabilities_tuples = [convert_probabilities(row) for row in probabilities]
max_prob_info = [max_probability_info(row) for row in probabilities]

pred_class, max_prob = zip(*max_prob_info)

df_dev = pd.DataFrame({
    'X': X_dev.tolist(),
    'y': y_dev.tolist(),
    'pred': pred_class,
    'max_prob': max_prob,
    'p': probabilities_tuples
})

# Correction for class 16 (Unknown, but not yet in training data) and class 17 (Wybren van Haga)
df_dev['pred'] = df_dev['pred'].apply(lambda x: 17 if x == 16 else x)
```

```python
known_probs = df_dev[df_dev['y'] != 16]['max_prob']
unknown_probs = df_dev[df_dev['y'] == 16]['max_prob']
```

```python
plt.hist(known_probs, bins=50, alpha=0.5, label='Politicians', color='skyblue')
plt.hist(unknown_probs, bins=50, alpha=0.5, label='Unknowns', color='magenta')
plt.xlabel('Maximum Probability')
plt.ylabel('Frequency')
plt.legend(loc='upper center')
plt.title('Distribution of Maximum Probabilities', fontdict=title_font)
plt.show()
```

```python
# Introduce threshold 
t1_threshold = 0.5
t2_threshold = 0.6
t3_threshold = 0.7
df_dev['pred_t1'] = df_dev.apply(lambda row: 16 if row['max_prob'] < t1_threshold else row['pred'], axis=1)
df_dev['pred_t2'] = df_dev.apply(lambda row: 16 if row['max_prob'] < t2_threshold else row['pred'], axis=1)
df_dev['pred_t3'] = df_dev.apply(lambda row: 16 if row['max_prob'] < t3_threshold else row['pred'], axis=1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
true_labels = df_dev['y']
predicted_labels_list = [df_dev['pred_t1'], df_dev['pred_t2'], df_dev['pred_t3']]
threshold_names = ['t1: 0.5', 't2: 0.6', 't3: 0.7']

results = []

for i, (predicted_labels, threshold_name) in enumerate(zip(predicted_labels_list, threshold_names), start=1):
    
    cm = confusion_matrix(true_labels, predicted_labels, labels=list(range(18)))

    accuracy_sc = accuracy_score(true_labels, predicted_labels)
    precision_sc = precision_score(true_labels, predicted_labels, average=None, zero_division=0)
    recall_sc = recall_score(true_labels, predicted_labels, average=None, zero_division=0)
    f1_sc = f1_score(true_labels, predicted_labels, average=None, zero_division=0)
    
    # Extract false negatives and true positives
    false_negatives = cm.sum(axis=1) - np.diag(cm)
    true_positives = np.diag(cm)
    
    # Calculate false negative rate
    false_negative_rate = false_negatives.sum() / (false_negatives.sum() + true_positives.sum())
    
    # Store the results
    results.append({
        'Threshold': threshold_name,
        'Accuracy': accuracy_sc,
        'Precision': precision_sc.mean(),  # Mean precision for overall view
        'Recall': recall_sc.mean(),        # Mean recall for overall view
        'F1': f1_sc.mean(),                # Mean F1 score for overall view
        'False Negative Rate': false_negative_rate
    })
    
    # Normalize the confusion matrix by row (true labels)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    # Create DataFrames for better visualization
    cm_df = pd.DataFrame(cm, index=list(range(18)), columns=list(range(18)))
    cm_normalized_df = pd.DataFrame(cm_normalized, index=list(range(18)), columns=list(range(18)))
    
    # Plot the normalized confusion matrix
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    sns.heatmap(cm_normalized_df, annot=True, fmt='.2f', cmap='Blues', ax=axes[0], cbar_kws={'label': 'Proportion'}, vmin=0, vmax=1)
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('True')
    axes[0].set_title(f'Normalized Confusion Matrix ({threshold_name})', fontdict={'fontsize': 14})
    
    # Plot the non-normalized confusion matrix with adjusted color scale
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', ax=axes[1], cbar_kws={'label': 'Count'}, vmax=np.percentile(cm, 95))
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('True')
    axes[1].set_title(f'Confusion Matrix with Adjusted Color Scale ({threshold_name})', fontdict={'fontsize': 14})
    
    plt.tight_layout()
    plt.show()

# Print the accuracy and false negative rate results
for result in results:
    print(f"Threshold metrics for {result['Threshold']}")
    print(f"Accuracy: {result['Accuracy']:.3f}")
    print(f"Precision: {result['Precision']:.3f}")
    print(f"Recall: {result['Recall']:.3f}")
    print(f"F1: {result['F1']:.3f}")
    print(f"False Negative Rate: {result['False Negative Rate']:.4f}\n\n")
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

true_labels = np.where(df_dev['y'] == 16, 1, 0)
predicted_labels_list = [np.where(df_dev['pred_t1'] == 16, 1, 0), np.where(df_dev['pred_t2'] == 16, 1, 0), np.where(df_dev['pred_t3'] == 16, 1, 0)]
threshold_names = ['t1: 0,5', 't2: 0,6', 't3: 0,7']

results = []

for i, (predicted_labels, threshold_name) in enumerate(zip(predicted_labels_list, threshold_names), start=1):
    
    cm = confusion_matrix(true_labels, predicted_labels, labels=list(range(2)))
    
    # Extract true positives, false positives, true negatives, and false negatives for class 16
    true_positives = cm[1, 1]
    false_positives = cm[0, 1]
    false_negatives = cm[1, 0]
    true_negatives = cm[0, 0]

    accuracy_class_16 = accuracy_score(true_labels, predicted_labels)
    precision_class_16 = precision_score(true_labels, predicted_labels, pos_label=1, zero_division=0)
    recall_class_16 = recall_score(true_labels, predicted_labels, pos_label=1, zero_division=0)
    f1_class_16 = f1_score(true_labels, predicted_labels, pos_label=1, zero_division=0)
    
    # Calculate false positive rate for class 16
    false_positive_rate_class_16 = false_positives / (false_positives + true_negatives)
    
    # Store the results
    results.append({
        'Threshold': threshold_name,
        'Accuracy Class 16': accuracy_class_16,
        'Precision Class 16': precision_class_16,
        'Recall Class 16': recall_class_16,
        'F1 Score Class 16': f1_class_16,
        'False Positive Rate Class 16': false_positive_rate_class_16
    })
    
    # Normalize the confusion matrix by row (true labels)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    # Create DataFrames for better visualization
    cm_df = pd.DataFrame(cm, index=[0, 1], columns=[0, 1])
    cm_normalized_df = pd.DataFrame(cm_normalized, index=[0, 1], columns=[0, 1])
    
    # Plot the normalized confusion matrix
    fig, axes = plt.subplots(1, 2, figsize=(8, 3))
    
    sns.heatmap(cm_normalized_df, annot=True, fmt='.2f', cmap='Blues', ax=axes[0], cbar_kws={'label': 'Proportion'}, vmin=0, vmax=1)
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('True')
    axes[0].set_title(f'Normalized Confusion Matrix ({threshold_name})', fontdict=title_font)
    
    # Plot the non-normalized confusion matrix with adjusted color scale
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', ax=axes[1], cbar_kws={'label': 'Count'}, vmax=np.percentile(cm, 95))
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('True')
    axes[1].set_title(f'Confusion Matrix ({threshold_name})', fontdict=title_font)
    
    plt.tight_layout()
    plt.show()


# Print the metrics results
for result in results:
    print(f"Threshold metrics for {result['Threshold']}")
    print(f"Accuracy (unknowns): {result['Accuracy Class 16']:.3f}")
    print(f"Precision (unknowns): {result['Precision Class 16']:.3f}")
    print(f"Recall (unknowns): {result['Recall Class 16']:.3f}")
    print(f"F1 Score (unknowns): {result['F1 Score Class 16']:.3f}")
    print(f"False Positive Rate (unknowns): {result['False Positive Rate Class 16']:.3f}\n\n")
```
